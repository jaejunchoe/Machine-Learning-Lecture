```python
import numpy as np
import pandas as pd
import io

#1
from google.colab import drive
drive.mount('content')
filename = '/content/iris.csv'
df = pd.read_csv(filename)

print(df.info(), '\n')
print(df.describe(), '\n')
```

```python
# 2
df_dep = df["variety"]                #종속변수
df_ind = df.drop(["variety"], axis=1) #독립변수

print(df_dep.info(), df_ind.info())

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(df_ind, df_dep, test_size = 0.3, random_state=20)

print(train_input.info())
print(test_input.info())
```

```python
# 3

# 모델1
from sklearn.preprocessing import StandardScaler
ss_scaler_1 = StandardScaler()
ss_scaler_1.fit(train_input)
ss_scaled_1 = ss_scaler_1.transform(train_input)

from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(n_estimators = 64, max_depth = 2, random_state = 20)
rfc_1.fit(ss_scaled_1, train_target)
```

```python
# 모델2
from sklearn.ensemble import RandomForestClassifier
rfc_2 = RandomForestClassifier(n_estimators = 64, max_depth = 2, random_state = 20)
rfc_2.fit(train_input, train_target)
```

```python
# 4
# 4.1
ss_scaled_test = ss_scaler_1.transform(test_input)

y_pred_scaled = rfc_1.predict(ss_scaled_test)
print(y_pred_scaled)

# 4.2
y_predict = rfc_2.predict(test_input)
print(y_predict)
```

```python
# 5
from sklearn.metrics import classification_report
classification_report(test_target, y_pred_scaled)
```

```python
classification_report(test_target, y_predict)
```
