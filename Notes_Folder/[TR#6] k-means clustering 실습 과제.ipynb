```python
import numpy as np
import pandas as pd
import io

# 0.‘home appliance_LG.csv’ 파일을 로드하여 X라 저장합니다. 이 데이터를 그대로 사용합니다. (train/test 분리 안함)
from google.colab import drive
drive.mount('content')
filename = '/content/home appliance_LG.csv'
X = pd.read_csv(filename)
```

```python
# 1. 어떤 변수가 있는지, 어떤 통계량을 갖는지, missing value는 없는지 등을 살펴봅니다.
 
print(X.info(), '\n')
print(X.describe(), '\n')
```

```python
# 2
A = X['Num_refrigerator']
B = X['Du_refrigerator']
C = B/A

x1 = C
x2 = A

df = pd.concat([x1,x2],axis=1)
print(df)

df.columns=['0','Num_refrigerator']
df.rename(columns = {'0': '요리 시 냉장고 문 1회 열었을때의 사용 시간', 'Num_refrigerator': '요리 시 냉장고 문 연 횟수'}, inplace = True)

print(df)
```

```python
# 3
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(df)
print(kmeans.cluster_centers_)
```

```python
# 4
from sklearn.metrics import silhouette_score
for i in range(2,7):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    print('k =', i, kmeans.cluster_centers_)
    print(silhouette_score(df, kmeans.labels_, metric='euclidean'))
```

```python
# 5  앞에서 고려한 2개 변수 (2번에서 얘기한) 외 다른 변수를 사용하여 클러스터링을 수행해보세요.
# 그리고 분석 결과 및 분석 결과의 활용 방안에 대해 Ideation 해보세요.

# 요리 시간대별(오전과 오후)에 따라 쿡탑의 버너 사용개수에 어떠한 영향을 미치는가

x11 = X['Time_hour']
x22 = X['Num_cooktop']

df1 = pd.concat([x11,x22],axis=1)
print(df1)

df1.columns=['Time_hour','Num_cooktop']
df1.rename(columns = {'Time_hour': '요리를 하는 시간', 'Num_cooktop': 'cooktop의 사용개수'}, inplace = True)

print(df1)

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
kmeans1 = KMeans(n_clusters=24)
kmeans1.fit(df1)
print(kmeans1.cluster_centers_)
print(silhouette_score(df1, kmeans1.labels_, metric='euclidean'))

```
